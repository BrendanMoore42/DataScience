```python
# Importing Libraries
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
import seaborn as sns

from pylab import rcParams
from tqdm import tqdm

# Set plotting format
#plt.rcParams['figure.figsize'] = (30,20)
%config InlineBackend.figure_format = 'retina'

# Disable notebook warnings
warnings.filterwarnings('ignore')


# Setting Dataframe format
pd.set_option('display.max_columns',1000)
#pd.set_option('display.float_format', lambda x: '%.1f' % x)
pd.set_option('precision',10)
sns.set()
```

```python
sample = pd.read_csv('Data/Leaf Classification_sample_submission.csv',
                     sep=',')
sample.head()
```

```python
test_data = pd.read_csv('Data/Leaf Classification_test.csv', 
                        sep=',', index_col='id')
test_data.head()
```

```python
train_data = pd.read_csv('Data/Leaf Classification_train.csv',
                         sep=',', index_col='id')
train_data.head()
```

```python
species_list = train_data.columns
```

```python
# Looking at dataframe shapes
train_data.shape, test_data.shape, sample.shape
```

```python
# Unique values for classification species
species_list = train_data.species.unique() 
pd.Series(species_list).count()
```

### Checking/Cleaning Null Values

```python
train_data.isnull().sum().sum()
```

```python
test_data.isnull().sum().sum()
```

### Converting Dataframe to Arrays

```python
# Converting dataframes to Arrays
X_train = train_data[train_data.columns[1:]].values
y_train = train_data.species.values
X_test = test_data.values
```

### Training the Model: KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

### Predicting Classification: KNN Classifier

```python
pd.Series(knn.predict(X_test), name='Prediction').head()
```

```python
submission = pd.DataFrame(knn.predict_proba(X_test), columns=[species_list])
submission['id'] = submission.index
submission = submission['id'] + submission[species_list]
submission.head()
```

### Exporting Results to CSV

```python
submission.to_csv('Submissions/leafclassification.csv')
```
