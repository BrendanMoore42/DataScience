### Strongest Pokemon Scorecards

```python
# Top 10 non-legendary pokemon
pok[pok['Legendary'] == True].sort_values('Total',ascending=False)[['Name','Total']].head(10)
```

```python
# Creating mage pokemont feature
pok['Mega']= pok.Name.str.lower().str.contains('mega')
```

```python
# Viewing the strongest non-legendary, mega pokemon
pok[(pok['Mega'] == True) & (pok['Legendary'] == False)].sort_values('Total',ascending=False)[['Name','Total']].head(10)
```

```python
# Top 10 non-legendary, non-mega pokemon
pok[(pok['Mega'] == False) & (pok['Legendary'] == False)].sort_values('Total',ascending=False)[['Name','Total']].head(10)
```



### The Generational Release of Pokemon

```python
# Wrangling a Dataframe that includes number of types, legendaries, and mega released by generation 
gen = pok.groupby('Generation')['Total','Type 1','Type 2'].count()
gen['Type 1'] = gen['Type 1'] - gen['Type 2']
gen.columns = ['Total','Single Type','Double Type']

# Adding on legendary only count to dataframe
leg_only = pok[(pok['Legendary'] == True)]
leg_only = leg_only[(leg_only['Mega'] == False)]
gen['Legendaries'] = leg_only.groupby('Generation')['Legendary'].count()

# Adding on megas only count to the dataframe (includes legendaries)
gen['Megas'] = pok[pok['Mega'] == True].groupby('Generation')['Mega'].count()
gen = gen.fillna(0)
gen
```

```python
# Wrangling Average Pokemon stats per generation release
pok.drop(['Name','Type 1','Type 2', 'Legendary','Mega'], axis=1).groupby('Generation').mean()
```
