### Creating numpy arrays for use with scikit-Learn

```python
X = train_df.drop(['Id','Cover_Type'],axis=1).values
y = train_df.Cover_Type.values
X_pred = test_df.drop('Id',axis=1).values

X.shape, y.shape, X_pred.shape
```

### Use crossvalidation to measure predictive power

```python
#Split into Traininng and Validation Sets
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size = 0.25)
```

```python
RFC = ensemble.RandomForestClassifier()
RFC.fit(X_train, y_train)
y_train_pred = RFC.predict(X_test)
```

```python
print (metrics.classification_report(y_test, y_train_pred))
print ()
print (metrics.accuracy_score(y_test, y_train_pred))
```

### Train with entire training set and predict results
