```python

import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
import numpy as np
import os
import pandas as pd
import datetime as dt



#Display max number of rows

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
#pd.set_option('display.max_row',1000)

#applications = '/Users/kavisekhon/Desktop/Applications.csv'
survey = '/Users/Kavi/Dropbox/Data/Datasets/Modified/Mental Health in Tech Survey/survey.csv'
survey = pd.read_csv(survey)
survey = pd.DataFrame(survey)
s = survey
s.head()
```

```python
#Selecting All Data within the first row indexed
s.loc[0,:]
```

```python
#Selecting All Data within multiple Rows - Method 1
s.loc[(0,1,2),:]
```

```python
#Selecting All Data within multiple Rows - Method 2
s.loc[0:2,:]
```

```python
#Selecting All Data within multiple Rows - Method 3 (Lazy)
s.loc[0:2]
```

```python
#Selecing All Data within a Column
s.loc[:,'Timestamp']
```

```python
#Selecing All Data within multiple Columns
s.loc[:,['Timestamp', 'Age','Gender']]
```

```python
#Selecing All Data within multiple Columns
s.loc[:, 'Timestamp':'Gender']
```

```python
# loc returning specified rows and columns

s.loc[1:5,'Timestamp':'Gender']
```

```python
# loc returning Gender Observations/Columns,when WHERE condition is met

s[s.Age == 44].Gender
```

```python
# loc returns all columns that meet where condition
s.loc[s.Age == 44, :]
```

```python
# Loc to Return columns, that meet Where condition
s.loc [s.Age == 44, 'Gender']
```

```python
# iloc is inclusive of the first number and exclusive of the second number
# iloc retunring all observation from the specified columns
s.iloc[:,[0,3]]
```

```python
#iloc returning all columns from specified observations
s.iloc[0:3,:]
```

```python
#Commonly Seen
s[['Age','Gender']] #Bad Habit
s.loc[:,['Age','Gender']] #Good Habit
```

```python
#Commonly Seen
s[0:2] #Bad
s.iloc[0:2,:] #Good
```

```python
#IX - Allows you to mix labels and integers
#allows selection by label and position
#ix was made in Pandas before loc or iloc
s.ix[0,'Age']


```

```python
# Both the number within the bracket [], can be a position or label

s.ix[0,1]
```

```python
s.ix[0:5,'Timestamp':'state']
```

```python
# Rows are done inclusively, and columns 2nd number are exclusive
s.ix[0:2,0:2]
```
