```python
x = [1,2,3,4,5,6]
def circular_arrays(x,n):
    matrix  =[]
    for i,j in enumerate(x):
        y = x[i:i+n]+x[:]
        counter = 0
        while len(y) < n:
            y = y+x[:]
        u = []
        for j in y:
            if counter == n:
                break
            #print(n)
            u.append(j)
            counter += 1
        #print(y)
        print(u)
        matrix.append(u)
    return matrix

for n in range(0,50):
    circular_arrays(x, n)
```
