### Random XKCD Comic Generator
Displays a Random XKCD article every 30 seconds

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(lines) {
    return false;
}
```

```python
import numpy as np
import re
import requests
import time


from IPython.display import Image, display
from IPython.display import clear_output


def retrieve_random_xcd_page():
    '''
    Generate a random number from 0 to 2021 and outputs a 
    andom comic url from xkcd
    
    Parameters
    ----------
    None
    
    Returns
    -------
    
    Example
    -------
    >>>> retrieve_random_xcd_page()
    >>>> url = retrieve_random_xcd_page()
         print(url)

    '''
    com_num = np.random.randint(0,2021)
    page = requests.get('https://xkcd.com/'+str(com_num)+'/')
    content = page.text
    search = re.findall('//imgs.xkcd.com/comics[^"]+', content)
    comic_info = search[0]
    skeleton = "https:"+str(comic_info)
    return skeleton 

  
# This wile loop will show use the comic for 20 seconds
while True:
    comic_url = retrieve_random_xcd_page()
    #print(comic_url)
    display(Image(comic_url))
    time.sleep(20)
    clear_output()
```
