# Import Data by Pandas
    
<span> If you starting from scratch and are unfamiliar with import data to notebooks. The pandas *pd.read_csv()* is the most powerful command that you will come across this one function work really well and contains over 40+ parameter to adjust how you import the incoming data. A cool feature of this function is that you can place dropbox links (or any public link to a CSV file) inside the function and you can download the data right from the internet.</span>
    
### Import Prelimnaries

```python
# Import Modules
import pandas as pd

pd.set_option('max_columns', 1000)
pd.set_option('max_rows', 1000)
```

### Import CSV

```python
# Read CSV Data
df = pd.read_csv('https://www.dropbox.com/s/5ui1and0j2dbdjh/pokemon.csv?dl=1')

# View the data
df.head(5)
```

### Import Excel Document

```python
pd.read_excel('https://www.dropbox.com/s/v704fefrr3r0neo/weather.xls?dl=1', Sheeet='weather')
```

### Import Text Data

```python
pd.read_csv('https://www.dropbox.com/s/q0ppmfbzlv0k9p4/ted_main.txt?dl=1', sep=" ")
```

### Import JSON Data

```python
pd.read_json('https://www.dropbox.com/s/16twozc11mwgwtk/sampledata.json?dl=1', orient='speakers')
```

### Import XML Data

```python
pd.read_xml('https://www.w3schools.com/xml/note.xml')
```
