# Group By Aggregations
### Import Preliminaries

```python
# Import modules
import pandas as pd

# Set pandas float display format
pd.options.display.float_format = '{:.1f}'.format

# Import data
df = pd.read_csv('https://www.dropbox.com/s/2uos48wjnkx4vcj/Pokemon.csv?dl=1',
                index_col=['#'])

# View the data
df.head(5)
```

### Pandas Describe Function

```python
df.describe()
```

### Series Groupby

```python
### Series group by pokemon generation via Total, show mean
gen = df.groupby('Generation').Total.mean()
gen = pd.DataFrame(gen); gen
```

### DataFrame Groupby

```python
# Dataframe group by on generation via Total, show mean
gen = df.groupby('Generation').mean()
gen = pd.DataFrame(gen); gen
```

### Multiple Group By Clause

```python
# series groupby on multiple categories (generation, legendary status) 
# via Attack, show max 

legendary = df.groupby(['Generation','Legendary']).Attack.mean()
legendary = pd.DataFrame(legendary); legendary
```

### Pivot Groupby Statement

```python
# Datafarme Groupby Statement on only Type 1 and Type 2 pokemon.
# Plus the use  Unstack function on the groupby
# View mean defense by Pokemon Type

ptype = df.groupby(['Type 1','Type 2']).Defense.mean().unstack()
ptype = ptype.fillna(value='-')
ptype = pd.DataFrame(ptype); ptype
```

### CrossTab Function

```python
# A count of only 2 type pokemon across their Types
two_type_pokemon_count = pd.crosstab(df['Type 1'], df['Type 2'])
two_type_pokemon_count
```
