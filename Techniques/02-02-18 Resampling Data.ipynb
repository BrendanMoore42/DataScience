# Resampling Data
### Import Preliminaries

```python
# Import modules
import pandas as pd

# Set pandas float display format
pd.options.display.float_format = '{:.1f}'.format

# Import data
df = pd.read_csv('https://www.dropbox.com/s/sx98q8x14nhd3xa/ww2_summary%20of%20weather.csv?dl=1',
                low_memory=False, usecols=['Date','MaxTemp','MinTemp','MeanTemp'],
                parse_dates=['Date'], index_col=['Date'])

# view the dataframe
df.sort_index(ascending=True).head()
```

```python
# Resample dataframe by Month
df.resample('M').mean().head(5)
```

```python
# Resample dataframe by Year
df.resample('Y').mean().head(5)
```

```python
# Resample dataframe for Every 2 Years
df.resample('2Y').mean().head(5)
```

```python
# Resample dataframe for 10 Days
df.resample('10D').mean().head(5)
```
