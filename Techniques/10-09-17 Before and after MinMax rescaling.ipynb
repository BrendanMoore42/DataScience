# Before and After MinMax Rescaling
### Import Modules

```python
%matplotlib inline

import pandas as pd
import seaborn as sns
from sklearn import preprocessing
```

###  Create DataFrame

```python
X = pd.DataFrame(data=[-500.5, -100.1, 0, 100.1, 900.9]).values
```

### Standardize Feature

```python
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))
rescaled_X = minmax_scale .fit_transform(X)
rescaled_X
```

### Visualize Orginal Feature

```python
sns.distplot(X)
```

### Visualize Standardized Feature

```python
sns.distplot(rescaled_X)
```
