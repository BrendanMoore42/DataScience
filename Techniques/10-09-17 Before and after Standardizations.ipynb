# Before and After Standardization
### Import Modules

```python
%matplotlib inline

import pandas as pd
import seaborn as sns
from sklearn import preprocessing
```

###  Create DataFrame

```python
X = pd.DataFrame(data=[-500.5, -100.1, 0, 100.1, 900.9]).values
```

### Standardize Feature

```python
scaler = preprocessing.StandardScaler()
standardized_X = scaler.fit_transform(X)
standardized_X
```

### Visualize Orginal Feature

```python
sns.distplot(X)
```

### Visualize Standardized Feature

```python
sns.distplot(standardized_X)
```
