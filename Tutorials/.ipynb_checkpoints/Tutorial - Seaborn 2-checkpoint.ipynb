```python
# Adding the "encoding = "ISO-8859-1" fixed the read_csv utf-8 error
df = pd.read_csv('pokemon.csv', encoding = "ISO-8859-1", index_col=0)
```

```python
df.head()
```

## Seaborn's Plotting Functions

```python
# recommende way
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
# removing regression line, and a hue to pokemon evolutions, 
sns.lmplot(x='Attack', y='Defense', data=df,
          fit_reg=False, 
          hue='Stage') 
```

## Customizing with Matplotlib

```python
# plot using seaborn
sns.lmplot(x='Attack', y='Defense', data=df,
          fit_reg=False, 
          hue='Stage') 
# Tweak using matplotlib
plt.xlim(0, None)
plt.ylim(0, None)
```

## The role of Pandas

```python
# default boxplot
sns.boxplot(data=df)
```

```python
# pre-format DataFrame

stats_df = df.drop(['Total','Stage','Legendary'], axis =1)

# new boxplot using stat_df
sns.boxplot(data=stats_df)
```

## Seaborn themes

```python
# set theme, then plot violin plot
sns.set_style('whitegrid')

# violin plot
sns.violinplot(x='Type 1', y='Attack', data=df)
```

## Color palettes

```python
# creating the pokemon colors
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                  ]
```

```python
# using customer color palette
sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette= pkmn_type_colors)
```

```python
# swarmplot with pokemon color palettes

sns.swarmplot(x='Type 1', y='Attack', data=df,
              palette= pkmn_type_colors)
```

## Overlaying plots

```python
# Set figuresize with matplot lib
plt.figure(figsize=(12,6))

# create a violin plot (inner = remove bars from inside violin)
sns.violinplot(x='Type 1', y='Attack', data=df,
              palette= pkmn_type_colors, inner=None)

# alpha = makes the plot slightly transparent, 'k' = black
sns.swarmplot(x='Type 1', y='Attack', data=df,
              color='k', alpha= 0.7)

# set title with matplotlib
plt.title('Attack by Type')
```

# Putting it all together

```python
stat_df = df
stat_df = stat_df.drop(['Legendary', 'Stage','Total'], axis =1)
stat_df.head()
```

```python
# melt dataframe
melted_df = pd.melt(stat_df,
                   id_vars=['Name', 'Type 1', 'Type 2'], 
                   var_name='Stat')
melted_df.head()
```

```python
# viewing the comparison
stat_df.shape, melted_df.shape
```

```python
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1')
```

```python
# swarmplot customizations
plt.figure(figsize=(12,6))

# slipt = seperate points by hue
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1', split=True, 
             palette = pkmn_type_colors)

# place the legend to the right
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

## Quick-Fire Data Visualizations (Pokedex Mini-Gallery

```python
# calculate correlations
corr = stat_df.corr()

# heatmap
sns.heatmap(corr)
```

```python
# histogram - df.Attack series datatype was string
sns.distplot(df.Attack)
```

```python
# countplot (a.k.a barplot)

sns.countplot(x='Type 1', data=df, palette = pkmn_type_colors)

# rotate x-labels
plt.xticks(rotation=45)
```

```python
# factor plot
g = sns.factorplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot

# rotate x-labels - ue set_xticklabels because there is a group
g.set_xticklabels(rotation=45)
```

```python
# Density plot

sns.kdeplot(df.Attack, df.Defense)
```

```python
# Joint Distribution Plot
sns.jointplot(x='Attack', y='Defense', data=df)
```
