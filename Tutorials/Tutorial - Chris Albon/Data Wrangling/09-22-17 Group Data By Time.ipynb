# Group Data by Time
### Preliminaries

```python
# Import required packages
import pandas as pd
import datetime
import numpy as np
```

### Create Data

```python
# Create a datetime variable for today
base = datetime.datetime.today()

# Create a list variable that creates 365 days of rows of datetime values
date_list = [base-datetime.timedelta(days=x) for x in range (0,365)]
```

```python
date_list
```

```python
# Create a list of varible for today
base = datetime.datetime.today()
# Create a list variable that creates 365 days of rows of datetime values
date_list = list(np.random.randint(low=1, high=1000, size=365))
```

```python
date_list
```

```python
# Cerea a list of variables for 365 numeric values
score_list = list(np.random.randint(low=1, high=1000, size=365))
```

```python
# Create an empty dataframe
df = pd.DataFrame()

# Create a column from the datetime variable
df['datetime'] = date_list

# Convert a column from the datetime variable
df['datetime'] = pd.to_datetime(df['datetime'])
# Set the datetime column as the index
df.index = df['datetime']

# Create a column from the numeric sorce variable
df['score'] = score_list
```

```python
# Let's take a look at the data
df.head()
```

### Group Data by Date

```python
# Group the data by month, and take the mean for each group (i.e. each month)
df.resample('M').mean()
```

```python
# Group the data by month, and take the sum for each group (i.e.each month)
df.resample('M').sum()
```
