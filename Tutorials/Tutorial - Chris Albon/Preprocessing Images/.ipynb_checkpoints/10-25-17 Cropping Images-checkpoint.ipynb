# Cropping Images
Source: https://chrisalbon.com/machine-learning/cropping_images.html
### Preliminaries

```python
# Load image
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Load Images As Greyscale

```python
# Load image as grayscale
image = cv2.imread('images/plane_256x256.jpg', cv2.IMREAD_GRAYSCALE)
```

### Crop Image

```python
# Select first half of the column and all rows
image_cropped = image[:,:126]
```

### View Image

```python
# View image
plt.imshow(image_cropped, cmap='gray'), plt.axis("off")
plt.show()
```
