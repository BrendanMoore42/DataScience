# Tag Parts of Speech
###  Preliminaries

```python
# Load libraries
from nltk import pos_tag
from nltk import word_tokenize
```

###  Create Text Data

```python
# Create Text
text_data = 'Chris loved outdoor running'
```

### Tag Parts of Speech

```python
# Use pre-trainined parts of speech tagger
text_tagged = pos_tag(word_tokenize(text_data))

# Show parts of speech
text_tagged
```
