### Inner Join

```python
%%sql

SELECT * FROM criminals
INNER JOIN crimes
ON criminals.pid=crimes.pid_arrested;
```

### Left Join

Returns all rows from the left table but only the rows from the right left that match the left table.

```python
%%sql

SELECT * FROM criminals
LEfT JOIN crimes
ON criminals.pid = crimes.pid_arrested;
```
