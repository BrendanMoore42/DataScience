### Missing No 

Source: https://github.com/ResidentMario/missingno

```python
import pandas as pd 
df = pd.read_csv('Data/h1b_kaggle.csv')
```

```python
df.isnull().sum()
```

```python
import missingno as msno
%matplotlib inline
msno.matrix(df.sample(1000))
```

```python
msno.bar(df.sample(1000))
```

```python
msno.heatmap(df)
```

```python
msno.dendrogram(df.sample(1000))
```
