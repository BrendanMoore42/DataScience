### Opening ZipFile Image

```python
import zipfile
from io import BytesIO
from PIL import Image

zf = zipfile.ZipFile('Data/messi1.zip', 'r')
print(zf.namelist())
```

```python
image_data = zf.read('messi1.jpg')
fh = BytesIO(image_data)
img = Image.open(fh)

plt.imshow(img); plt.show()
```
