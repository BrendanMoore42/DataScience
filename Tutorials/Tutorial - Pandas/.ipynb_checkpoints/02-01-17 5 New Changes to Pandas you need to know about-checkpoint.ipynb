### 1) ix is deprecated

```python
# read the drinnk dataset into a dataframe
drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col='country')
drinks
```

```python
# loc acccess by label
drinks.loc['Angola', 'spirit_servings']
```

```python
# iloc access by position
drinks.iloc[4,1]
```

```python
# ix access by label or position (newly deprecated)
drinks.ix['Angola',1]
```

```python
# alternative: use loc
drinks.loc['Angola', drinks.columns[1]]
```

```python
# alternative: use iloc
drinks.iloc[drinks.index.get_loc('Angola'), 1]
```

```python
# ix accesss by label or position (newly deprecated)
drinks.ix[4, 'spirit_servings']
```

```python
# alternative: use loc
drinks.loc[drinks.index[4], 'spirit_servings']
```

```python
#alternative: use iloc
drinks.iloc[4, drinks.columns.get_loc('spirit_servings')]
```
