# Barplots
    
<span>Plotting barplots for distributions is a common task in every dataset. The basic blue barplots can be boring and dull in matplotlib. I attempt to summarize the common method to plotting them across a few common python packages. Let's see how nice and crazy we can make our barplots with the current pacakges available to us.</span>

### Import Preliminaries

```python
# Import modules
import pandas as pd
import matplotlib as plt
df = pd.read_csv('Data/undergraduate_package.csv')

# View the data
df.head()
```

### Normal Barplot

```python
%matplotlib inline

### Plot Expenses by Year
data = df.groupby(['academic.year'])['cost'].sum()
data.plot.bar(color='black');
```

### Horizontal Barplot

```python
data = df.groupby(['academic.year'])['cost'].sum()
data.plot.barh(color='black', figsize=(5,10));
```

### Stacked Barplot

```python
data = df.groupby(['academic.year','component'])['cost'].sum()
data = data.unstack('component')
data  = data.drop('Total', axis=1)
data.plot.bar(stacked=True);
```
